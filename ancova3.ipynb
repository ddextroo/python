```python
from pingouin import ancova
import pandas as pd
import numpy as np
data3 = pd.DataFrame({'methodology': np.repeat(['Low-Carb', 'Mediterranean', 'Vegan'], 4),
                    'pre_diet_weight': [160, 170, 155, 165,
                                      170, 175, 160, 180,
                                      175, 180, 165, 170],
                    'post_diet_weight': [150, 160, 145, 155,
                                       155, 160, 150, 165,
                                       160, 165, 155, 160]})
data3
```

```python
ancova(data=data3, dv='post_diet_weight', covar='pre_diet_weight', between='methodology')
```
