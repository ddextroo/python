```python
from pingouin import ancova as anc1
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data1 = pd.DataFrame({'methodology': np.repeat(['Workshops', 'Online Courses', 'On-the-job'], 4),
					'pre_training_score': [78, 85, 76, 88,
									82, 79, 75, 84,
									85, 83, 79, 82],
					'post_training_score': [88, 90, 84, 87,
									91, 81, 86, 90,
									92, 87, 88, 89]})
data1
```

```python
result = ancova(data=data1, dv='post_training_score', covar='pre_training_score', between='methodology')
ancova(data=data1, dv='post_training_score', covar='pre_training_score', between='methodology')
```

```python
# Create a bar chart to present ANCOVA results
plt.figure(figsize=(10, 6))
sns.barplot(data=result, x='Source', y='F', hue='Source', dodge=False, palette='viridis')
plt.xlabel('Factors')
plt.ylabel('F-statistic')
plt.title('ANCOVA Results')
plt.yscale('log')  # Use log scale for the y-axis for better visualization of p-values

# Add p-values as text labels above the bars
for i, row in result.iterrows():
    plt.text(i, 1.1 * row['F'], f'p = {row["p-unc"]:.4f}', ha='center', fontsize=10)

# Remove the legend and adjust plot aesthetics
plt.legend([], frameon=False)
plt.grid(axis='y', linestyle='--', alpha=0.6)

# Show the plot
plt.tight_layout()
plt.show()
```
