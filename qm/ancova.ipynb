```python
from pingouin import ancova as anc1
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data1 = pd.DataFrame({'methodology': np.repeat(['Workshops', 'Online Courses', 'On-the-job'], 4),
					'pre_training_score': [78, 85, 76, 88,
									82, 79, 75, 84,
									85, 83, 79, 82],
					'post_training_score': [88, 90, 84, 87,
									91, 81, 86, 90,
									92, 87, 88, 89]})
data1
```

```python
result = ancova(data=data1, dv='post_training_score', covar='pre_training_score', between='methodology')
ancova(data=data1, dv='post_training_score', covar='pre_training_score', between='methodology')
```
