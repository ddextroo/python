```python
from pingouin import ancova
import pandas as pd
import numpy as np
data2 = pd.DataFrame({'methodology': np.repeat(['Social Media Ads', 'Email Marketing', 'TV Commercials'], 4),
                    'pre_campaign_sales': [12000, 14000, 11000, 13000,
                                          13500, 12500, 11500, 14000,
                                          14200, 13000, 12000, 13800],
                    'post_campaign_sales': [15000, 16000, 13000, 14500,
                                           15500, 14500, 13500, 16000,
                                           16000, 15000, 14000, 15500]})

data2
```

```python
ancova(data=data2, dv='post_campaign_sales', covar='pre_campaign_sales', between='methodology')
```
